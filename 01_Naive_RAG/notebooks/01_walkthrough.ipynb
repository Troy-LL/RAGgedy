### 1. Document Loading
A RAG pipeline relies on your custom context. We have a set of generated `.txt` passage files sitting in our data folder. Let's peek at them:

```python
from llama_index.core import SimpleDirectoryReader

documents = SimpleDirectoryReader(config.PASSAGES_DIR).load_data()
print(f"Loaded {len(documents)} documents! Here is the first one's content:\n")
print(documents[0].text)
```

### 2. Chunking
If we fed the entire corpus at once, context windows would overflow, and LLM attention drops making accuracy worse. We must break documents down into chunks.

Notice how we set **Chunk Size** to 512 tokens and **Overlap** to 64 to keep sentences intact.

```python
from llama_index.core.node_parser import SentenceSplitter

parser = SentenceSplitter(chunk_size=config.CHUNK_SIZE, chunk_overlap=config.CHUNK_OVERLAP)
nodes = parser.get_nodes_from_documents(documents)
print(f"Broke {len(documents)} documents into {len(nodes)} overlapping nodes/chunks.")
```

### 3. Embeddings & Indexing
To retrieve these text chunks without standard keyword-matching limitations, we convert their meaning into high-dimensional vectors (Embeddings) using `BAAI/bge-m3`.
